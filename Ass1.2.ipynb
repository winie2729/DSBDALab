```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
n = pd.read_excel("/content/Demo Data DSBDA (1).xlsx")
```

```python
d = n.filter(items = ['DM'])
```

```python
d
```

```python
d['DM'] = pd.to_numeric(d['DM'], errors = 'coerce')



```

```python
d =d.dropna(subset = ['DM'])
```

```python
fig = plt.figure(figsize = (10,7))
plt.boxplot(d['DM'])

fig.suptitle('Boxplot for DM')
plt.show()

```

```python
plt.hist(d['DM'], bins=5, color=['magenta'])
plt.title('Histogram for DM')
plt.show()
```
