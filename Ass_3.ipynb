```python
#Q 2 Write a Python program to display some basic statistical details like percentile, mean, standard
#deviation etc. of the species of ‘Iris-setosa’, ‘Iris-versicolor’ and ‘Iris- versicolor’ of iris.csv
#dataset.
import numpy as np
```

```python
import pandas as pd
csv_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
col_names = ['Sepal_Length','Sepal_Width','Petal_Length','Petal_Width','Species']
d = pd.read_csv(csv_url,names = col_names)
d
```

```python
#mean
print("The mean of Sepal Length is :",d.loc[:,"Sepal_Length"].mean())
```

```python
#median
print("The median of Sepal Length is :",d.loc[:,"Sepal_Length"].median())
```

```python
#mode
print("The mode of Sepal Length is :",d.loc[:,"Sepal_Length"].mode())
```

```python
#SD
print("The mean of Sepal Length is :",d.loc[:,"Sepal_Length"].std())
```

```python
#Storing Sepal Length and Sepal Width in array and then finding maximum and minimum values
array1 = np.array(d['Sepal_Length'])
array2=np.array(d["Sepal_Width"])
print("Sepal Length arr ",array1)
print("Sepal Width Arr ",array2)
```

```python
print("Maximum value of sepal length is :",max(array1))
print("Minimum value of sepal length is :",min(array1))
print("Maximum value of sepal width is :",max(array2))
print("Minimum value of sepal length is :",min(array2))
```

```python
# Replacing the categorical values by numeric values

d['Sepal_Length'].replace({5.1: 5.0}, inplace=True)
d['Sepal_Width'].replace({3.5: 3.6}, inplace=True)
d.head()
```

```python
d["Petal_Length"].replace({ "Yes":1, "No":0}, inplace=True)
d.head()
```
